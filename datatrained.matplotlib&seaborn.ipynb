```python
sal=[15,22,24,25,30,23,34,43,19,20,52,21]
sal.sort()
sal
```



Please see below example,

[15, 19, 20,21, 22, 23, 24, 25, 30, 34, 43, 52]

n total elements in the above exampie (12 items)

25% = (25/100)(n+1) = 3.25 (data at 3.25 position is the 25% percentile. Let's say around 20.50)

50%(50/100)(n+1) = 6.5 (data at this position)

75% (75/100) (n+1) = 9.75 (data at this position

# OUTLIERS

```python
import pandas as pd
ser=pd.Series(sal)
ser
```

```python
# 1st quantile
q1=ser.quantile(0.25)
print('Q1',q1)
#3rd quantile
q3=ser.quantile(0.75)
print('Q3',q3)

# find the IQR (inter quantile range) to identify outliers

iqr=q3-q1
print('IQR',iqr)
```

# outlier detection formula

# higher side==> Q3+(1.5*IQR)

# Lower side==> Q1-(1.5*IQR)


```python
upside=q3+(1.5*iqr)
print(upside)
```

anything above 46.375 is outlier

```python
downside=q1-(1.5*iqr)
print(downside)
```

anything below 5.375 is outlier

# 4.matplotlib pi plot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load dataset from seaborn
mileage=sns.load_dataset('mpg')
mileage.head()
```

```python
avg_mpg=mileage.groupby(by='cylinders')[['mpg','cylinders']].mean()
avg_mpg
```

```python
# pie
plt.pie(x=avg_mpg['mpg'],labels=avg_mpg['cylinders'])
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(x=avg_mpg['mpg'],labels=avg_mpg['cylinders'],explode=[0,0,0,0.05,0.3],autopct='%.2f')
plt.title('Avg_miles_per_cylinder_type',fontsize=20)
plt.show()
```

```python
# another example
rupees_monthly=[2000,2530,900,15000,1500]
items=['Tution_fee','Mobile_bill','Electricity_bill','Rent','Misc']
df=pd.DataFrame({'Expense':rupees_monthly,'items':items})
df
```

```python
plt.figure(figsize=(8,8))
plt.pie(x=df['Expense'],labels=df['items'],explode=(0,0,0,0.05,0.3),autopct='%.3f',rotatelabels=True)
plt.show()
```
